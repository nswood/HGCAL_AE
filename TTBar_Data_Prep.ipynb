```python
def prep_ttbar(n = 0):
    arr = torch.tensor(pd.read_csv('Ecoder/TrainingData/more_testing_1.csv').to_numpy(),dtype = torch.float)
    if n > 0:
        for i in range(n):
            arr = torch.vstack([arr,torch.tensor(pd.read_csv(f'Ecoder/TrainingData/more_testing_{i+2}.csv').to_numpy(), dtype = torch.float)])
    return arr
```

```python
dt = prep_ttbar(80)
```
